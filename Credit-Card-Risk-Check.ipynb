## Credit Card Risk Check

```python
import pandas as pd
import numpy as np
```

```python
credit_df=pd.read_csv('Credit_dataset.csv')
credit_df.head()
```

```python
#removing id column since it is not required to give the required output
credit_df=credit_df.drop(['ID'],axis=1)
credit_df.head()
```

```python
#renaming PAY_0 to PAY_1 to build a pattern
credit_df.rename(columns={'PAY_0':'PAY_1'}, inplace=True)
credit_df.head()
```

```python
#we need to remove unwanted categories in features like education and marriage
credit_df['EDUCATION'].value_counts()
credit_df['MARRIAGE'].value_counts()
```

```python
#mapping categories 0,5,6 of education to category 4 since category 0,5,6 are not defined.1->school,2->high school,3->College,4->university
credit_df['EDUCATION']=credit_df['EDUCATION'].map({0:4,1:1,2:2,3:3,4:4,5:4,6:4})
credit_df['MARRIAGE']=credit_df['MARRIAGE'].map({0:3,1:1,2:2,3:3})
credit_df.head()
```

```python
##Standar scaling
from sklearn.preprocessing import StandardScaler
scaling=StandardScaler()
x=credit_df.drop(['default.payment.next.month'],axis=1)
x=scaling.fit_transform(x)
y=credit_df['default.payment.next.month']
```

### Applying xgboost

```python
#Hyperparameter optimization
params={
 "learning_rate"    : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,
 "max_depth"        : [ 3, 4, 5, 6, 8, 10, 12, 15],
 "min_child_weight" : [ 1, 3, 5, 7 ],
 "gamma"            : [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
 "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ]
    }
```

```python
#Hyperparameter optimization using RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
import xgboost
```

```python
def timer(start_time=None):
    if not start_time:
        start_time = datetime.now()
        return start_time
    elif start_time:
        thour, temp_sec = divmod((datetime.now() - start_time).total_seconds(), 3600)
        tmin, tsec = divmod(temp_sec, 60)
        print('\n Time taken: %i hours %i minutes and %s seconds.' % (thour, tmin, round(tsec, 2)))
```

```python
classifier=xgboost.XGBClassifier()
```

```python
random_search=RandomizedSearchCV(classifier,param_distributions=params,n_iter=5,scoring='roc_auc',n_jobs=-1,cv=5,verbose=3)
```

```python
from datetime import datetime
# Here we go
start_time = timer(None) # timing starts from this point 
random_search.fit(x,y)
timer(start_time) # timing ends here 
```

```python
random_search.best_estimator_
```
